````python
def get_target_names(x):
    if x == 0:
        return "setosa"
    if x == 1:
        return "versicolor"
    if x == 2:
        return "virginica"
````

````python
a = pd.DataFrame(columns=iris_dataset['feature_names'],data = iris_dataset['data'])
a['target'] = iris_dataset['target']
a["target_names"] = a['target'].apply(lambda x : get_target_names(x))
a.head(10)
````

````python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(
    a.loc[:,['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']],
    a['target'],
    random_state=0
) 
print(X_train.head(2))
print(X_train.shape)
print(X_test.head(2))
print(X_test.shape)
print(y_train.head(2))
print(y_train.shape)
print(y_test.head(2))
print(y_test.shape)
````
