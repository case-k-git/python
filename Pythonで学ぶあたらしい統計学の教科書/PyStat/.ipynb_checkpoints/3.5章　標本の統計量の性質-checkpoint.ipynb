````python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
%precision 3
````

````python
#母集団
population = stats.norm(loc = 4, scale = 0.8)
````

````python
sample_mean_array = np.zeros(10000)
sample_mean_array[0]
````

````python
# 乱数の種類
np.random.seed(1)
#繰り返し
for i in range(0, 10000):
    #標本抽出　& 平均値
    sample = population.rvs(size = 10)
    sample_mean_array[i] = sp.mean(sample)
````

````python
sample_mean_array
````

````python
# 標本平均の平均値は母平均に近い
sp.mean(sample_mean_array)
````

````python
# 標本平均の標準偏差
sp.std(sample_mean_array, ddof = 1)
````

```
母標準へんんさは0.8だったのでそれよりも小さくなっていることが
確認できる。
```

````python
sns.distplot(sample_mean_array, color = 'black')
````

````python
size_array = np.arange(start = 10, stop = 100100, step = 100)
size_array
````

````python
sample_mean_array_size = np.zeros(len(size_array))
````

````python
# 乱数の種類
np.random.seed(1)
#繰り返し
for i in range(0, len(size_array)):
    #標本抽出　& 平均値
    sample = population.rvs(size = size_array[i])
    print(len(sample))
    sample_mean_array_size[i] = sp.mean(sample)
    print(sample_mean_array_size[i])
````

````python
plt.plot(size_array, sample_mean_array_size, color = 'black')
plt.xlabel("sample size")
plt.ylabel("sample mean")
````

```
サンプルサイズ(母集団から抽出した標本数)が大きくなるほど、標本平均は
母平均に近づく。分散がなくなる
```

````python
# 標本平均を何度も計算する関数
def calc_sample_mean(size, n_trial):
    #リスト：試行回数分
    sample_mean_array = np.zeros(n_trial)
    for i in range(0, n_trial):
        sample = population.rvs(size = size)
        sample_mean_array[i] = sp.mean(sample)
    return(sample_mean_array)
````

````python
np.random.seed(1)
sp.mean(calc_sample_mean(size = 10,n_trial = 10000))
````

````python
# サンプルサイズを変更した時の標本平均の分布
np.random.seed(1)
# サンプルサイズ10 np.tile(a,reps) a:並べたい配列, reps：どの軸方向に何回繰り返すのか
size_10 = calc_sample_mean(size = 10, n_trial = 10000)
size_10_df = pd.DataFrame({
    'sample_mean':size_10,
    'size':np.tile("size = 10",10000)
})
# サンプルサイズ20
size_20 = calc_sample_mean(size = 20, n_trial = 10000)
size_20_df = pd.DataFrame({
    'sample_mean':size_20,
    'size':np.tile("size = 20",10000)
})
# サンプルサイズ30
size_30 = calc_sample_mean(size = 30, n_trial = 10000)
size_30_df = pd.DataFrame({
    'sample_mean':size_30,
    'size':np.tile("size = 30",10000)
})
# 結合
sim_result = pd.concat([size_10_df, size_20_df, size_30_df])
#　結果の表示
print(sim_result.head())
````

````python
#バイオリンプロット：サンプルサイズ別の標本平均の分布
sns.violinplot(x = "size", y = "sample_mean", data = sim_result, color = 'gray')
````

```
気づき：
サンプルサイズが大きくなると標本平均のばらつきは小さくなり、
母平均の近くに集中する。
```

# 標本平均の標準偏差は、母標準差よりも小さい
```
サンプル合図が大きくなると、標本平均の標準偏差は小さくなることを
証明する。
```

````python
# サンプルサイズ：2-100まで2区切り
size_array  = np.arange(start = 2, stop = 102, step = 2)
size_array
````

````python
#標本平均の標準偏差を格納する入れ物
sample_mean_std_array = np.zeros(len(size_array))

#シュミレーションの実行
np.random.seed(1) 
for i in range(0, len(size_array)):
    #標本平均
    sample_mean = calc_sample_mean(size = size_array[i], n_trial = 100)
    #標準偏差
    sample_mean_std_array[i] = sp.std(sample_mean, ddof = 1)
plt.plot(size_array, sample_mean_std_array, color = 'black')
plt.xlabel("sample size")
plt.ylabel("mean_std value")
````

```
サンプルサイズが増えると、標本平均の標準偏差が小さくなることが確認できる。
→　サンプルサイズを大きくすると、ばらつきの少ない信頼できる標本平均が得られる。
```

# 標準誤差

````python
standard_error = 0.8 / np.sqrt(size_array)
standard_error
````

```
標準誤差とシュミレーションの結果を重ね合わせる
```

````python
plt.plot(size_array, sample_mean_std_array, color = 'black')
plt.plot(size_array, standard_error, color = 'black',linestyle = 'dotted')
plt.xlabel("sample size")
plt.ylabel("mean_std_value")
````

````python
# リスト：標本分散
sample_var_array = np.zeros(10000)
np.random.seed(1)
for i in range(0,10000):
    sample = population.rvs(size = 10)
    sample_var_array[i] = sp.var(sample, ddof = 0)
sp.mean(sample_var_array)
````

```
母分散は0.8の２乗で0.64となるはず。
0.575となっており、分散を過小評価している。
```

# 不偏分散を使うとバイアスがなくなる。

````python
# リスト：標本分散
unbias_var_array = np.zeros(10000)
np.random.seed(1)
for i in range(0,10000):
    sample = population.rvs(size = 10)
    unbias_var_array[i] = sp.var(sample, ddof = 1)
sp.mean(unbias_var_array)
````

```
正しい、『0.64』付近にあることがわかる。不偏分散の平均値は母分散とみなしても良さそう。
```

# サンプルサイズ大なら、不偏分散は母分散に近い
```
・対象：不偏分散
・変化させるもの：サンプルサイズ
・調べたいこと：サンプルサイズが大きくなると、不偏分散は母分散に近づいていく
```

````python
# サンプルサイズ
size_array = np.arange(start = 10, stop = 100100, step = 100)
size_array
````

````python
# 不偏分散
unbias_car_array_size = np.zeros(len(size_array))
````

````python
np.random.seed(1)
for i in range(0, len(size_array)):
    sample = population.rvs(size = size_array[i])
    unbias_car_array_size[i] = sp.var(sample, ddof = 1)
plt.plot(size_array, unbias_car_array_size,color = 'black')
plt.xlabel("sample size")
plt.ylabel("unbias var")
````

```
このグラフからサンプルサイズが大きくなるにつて
不偏分散は0.64に近づいていく。
```

# 中心極限定理
```
母集団分布がなんであっても、サンプルサイズが大きい時には、確率変数の
和は正規分布に近いものとなる。これを中心極限定理と呼ぶ
```

````python
# コインを10000回投げた時の表になった回数
n_size = 10000
n_trial = 50000
#表なら1,裏ならば0を返す
coin = np.array([0,1])
#表が出た回数
count_coin = np.zeros(n_trial)
#コインをn_size回投げる思考をn_trial回行う。
np.random.seed(1)
for i in range(0,n_trial):
    count_coin[i] = sp.sum(
        np.random.choice(coin, size = n_size, replace = True)
    ) 
#ヒストグラム
sns.distplot(count_coin, color = 'black')
````
