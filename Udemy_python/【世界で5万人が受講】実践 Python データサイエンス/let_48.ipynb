```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from numpy.random import randn
```

```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
```

```python
dataset2 = randn(80)
```

```python
plt.hist(dataset2, color="indianred")
```

```python
#全体を積分(足し算)したときに１になるようにする
plt.hist(dataset1,normed=True)
```

```python
#ヒストグラムを重ねて書いて比較させることができる
plt.hist(dataset1,normed=True,bins=20)
plt.hist(dataset2,normed=True,bins=20, color="indianred")
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
#同時分布の可視化
sns.jointplot(data1, data2)
```

```python
#2次元のヒストグラムとなる
#色の濃さから全体的にどこにデータが多く分布しているかわかる。
sns.jointplot(data1, data2,kind="hex")
```
