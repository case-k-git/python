# パラメータとは
```
機械学習ではハイパーパラメータという。
モデルが複雑になるとこのパラメターが増える。
ex 車
タイヤの大きさは？車の高さなど。
スクリーンショット 2018-02-20 9.56.57

決定木ではなんのパラメータを選択すればいいのか？
決定木では
①max_depth(決定木の深さ)
②min_samples_leaf(葉に属する最小サンプル数)

①max_dept
一番上を根、一番下を葉
この根〜葉が。深いほど細かいが過学習になる。

②最低のサンプル数
サンプル数が少なすぎると信憑性がなくなる

□　パラメータの検証方法
→　答えがわかっているデータ(学習データ)から仮の道のデータを作る
→　学習データをさらに分割(構築/検証)する
→　分割させて、パラメータを作る。検証データで精度が良いパラメータを見つける

□　どうやって良いパラメータを見つける
→　構築データと検証データの精度を比較する。
→　両方のデータの精度の乖離が小さく、検証データの良いパラメータを選ぶ
→　図に示されているようなグラフ(構築データ/検証データ)の乖離が小さいところが
最適なパラメータ。構築データは右上がり、検証データは一定の値で下がり始める。

□　具体的な分割方法
①　ホールドアウト
-単純な方法であるが欠点がある
-データが大量にあれば有効
②クロスバリデーション
-K分割交差検証とも言われる
-ホールドアウトの欠点を補った手法
-計算に時間がかかる

□　概要
①ホールドアウト法
・学習データを単純に2つに分割する
- 片方をモデル構築データとし、残りをモデル検証データとする
- 分割する割合は好きなものでう良い(通例、学習：検証=7:3)
- どこで分割するかで検証精度が大きく変わることもあり、不安定
ex 構築データが多すぎる
→　検証データが不十分となり検証不足
構築データが少なすぎる
→　モデルの学習不足

どこで分割するかが課題となる。

②クロスバリデーション
データをk個数に分割し、k回検証を行う。
ex 
K = 3の場合
①学習データを3個に分割する
②k-1個を構築データ、残り1個を検証データとする
全ての組み合わせを行う。K=3の場合は3通りの組み合わせがある。
各組み合わせに対して、精度を出す。
最終的に平均化させることで、構築データ、検証データの精度とする
→　データにばらつきがあっても解決される。
→　過学習になりにくいパラメータを見つけられる
```
