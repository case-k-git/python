```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#ヒストグラム
df = pd.read_csv('./data/sales.csv',dtype={'user_id': 'object', 'item_cd': 'object'},parse_dates=['date'])
plt.hist(df['amount'], bins=50, range=(0, 10000))
plt.title
plt.xlabel('Amount os sales')
plt.ylabel('Frequency')
```

```python
#ヒストグラム
plt.hist(df['amount'], bins=100)
plt.title('Frequency by amount of sales')
plt.xlabel('Amount of sales')
plt.ylabel('Frequency')
```

```python
#棒グラフ
df1 = pd.read_csv('./data/sales.csv',dtype={'user_id': 'object', 'item_cd': 'object'},parse_dates=['date'])
plt.hist(df1['amount'], bins=50, range=(0, 10000))
plt.title
index = df1.index
label = df1['dow']
plt.bar(index, df1['count'], tick_label=label)
plt.xlabel('Dow')
plt.ylabel('Frequency of purchases')
```
