````python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
print("KEY:\n{}".format(iris_dataset.keys()))
````

````python
print("value:\n{}".format(iris_dataset.values()))
````

````python
iris_dataset['data']
````

````python
iris_dataset['target_names']
````

````python
#サンプルの特徴量を可視化する。150レコード4種類のデータ
iris_dataset['data'].shape
````

````python
#p18 モデルの検証を行う
"""
データを2分割し、訓練データとテストデータに分けて利用する
iris_dataset['data']：ガクの長さ、幅、花弁の長さ、幅など
iris_dataset['target']：花の種類
"""
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
#train_test_split関数：擬似乱数を用いてデータをシャフルする
````

````python
#出力
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
````

````python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
#iris_dataset['feature_names']：特徴量の説明
````

````python
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker= 'o',
                       hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
````

````python
#K-最近傍方
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
#訓練
knn.fit(X_train, y_train)
````

````python
#予測を行う
X_new = np.array([[5, 2.9, 1,0.2]])
#サンプル1つの特徴量4つ
X_new.shape
````

````python
prediction = knn.predict(X_new)
iris_dataset['target_names'][prediction]
````

````python
#モデルの評価
"""
テストセットを活用したモデルの評価を行う。
"""
y_pred = knn.predict(X_test)
y_pred 
````

````python
knn.score(X_test, y_test)
"""
このモデルはテストデータに対して97%正しい
""
````
