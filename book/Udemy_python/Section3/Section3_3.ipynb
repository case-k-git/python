```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## データの読み込み
- train.csvを読み込みましょう
- 読み込んだものは変数trainに代入しましょう

```python
train = pd.read_csv("../case1/train.csv")
```

## trainの先頭行を確かめてみましょう

```python
train.head()
```

## trainのyの折れ線グラフを描いてみましょう
- 折れ線グラフを描くにはplot関数を使います

```python
train['y'].plot()
```

## グラフの大きさをもっと大きくしてみましょう
- カッコの中にオプションとしてfigsize=(12,4)と書きます

```python
train.y.plot(figsize=(12, 4))
```

```python
"""
気づき
⑴　時間が経過するごとにお客さんの売上げが下がっている
②　スパイクしている
時間とは別で憂機が変化する
"""
```

## グラフにタイトルをつけてみましょう
- カッコの中にオプションとして、title="XXX"と書きます
- カッコの中が複数となった場合はカンマ（,）を使うことで複数書くことができます

```python
train['y'].plot(title="y")
```

## グラフのx軸とy軸に名前を付けてみましょう
- ちょっと複雑で一旦、plot関数の結果を別の変数に代入する必要があります
- 今回は変数axに代入することにします
- x軸はaxを使い、set_xlabel("name")と書くことで名前をつけることができます

## trainのyのヒストグラムを描いてみましょう
- グラフを描く前に簡単にヒストグラムとは何かを理解しましょう
- plot.hist関数を使います

```python
"""
ヒストグラム
横軸　階級(横幅数値)
縦幅　度数(数値に該当するもの)
データがどのような形状をしているのか読み取ることができる。
①山が一つ：正規分布/ガウス分布
(平均が一番度数が大きい)
②山が2つある場合
ex テストでできる人、できない人がいる
2つある場合、2つのセグメントがあると判断することができる。
③山が偏っている(ロングテール)
データの分布と平均値に大きなズレがある。
④外れ値
100　より大きいところにある。外れ値を放置しておくと
想定と全く異なる挙動をする場合がある。見つけた上で何らかの処理が必要
ヒストグラムで可視化させることが大切
"""
```

```python
train['y'].plot.hist()
```

## グリッド線を引いてみましょう
- カッコの中にオプションとして、grid=Trueと書きます

```python
train['y'].plot.hist(grid=True)
```

## ヒストグラム上に平均値を表す赤線を描きましょう
- ヒストグラムで一番数が多いところが平均値とは限らないケースもある為、ヒストグラムに平均値を表す線を描いてみましょう
- まず縦線の描き方から始めます。縦線を描くときにはplt.axvline(x=NUMBER, color="red")と書きます
- vlineのvはvertical（垂直）

```python
plt.axvline(x=2,color="red")
```

```python
#xにyの平均値を当てはめる
plt.axvline(x=train['y'].mean(),color="red")
train['y'].plot.hist()
```

## 上記のヒストグラムを大きさ(12,4)にし、タイトルを"ヒストグラム"として、pngファイルの図として保存しましょう
- plt.savefig("FILE_NAME.拡張子")を一番最後に書きます
- ファイル名は「sample_fig.png」としましょう

```python
plt.axvline(x=train['y'].mean(), color="red")
train['y'].plot.hist(figsize=(12,4), title="ヒストグラム")
plt.savefig("sample_fig.png")
```

## yとweekの関係を調べる為に、箱ひげ図を描いてみましょう
- 箱ひげ図を描く前に簡単に箱ひげ図とは何かを理解しましょう
- 箱ひげ図の描き方は、yとweekを選択した上で、boxplot(by="week")と書きます
- by="week"はx軸にしたいカラムを選択します
- ○は外れ値です

```python
train[["y","week"]].boxplot(by="week")
```
