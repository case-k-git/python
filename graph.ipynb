````python
# 数値計算に使うライブラリ
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()

# 表示桁数の指定
%precision 3
# グラフを描画するライブラリ
from matplotlib import pyplot as plt

# グラフをjupyter Notebook内に表示させるための指定
%matplotlib inline
````

````python
x = np.array([9,10,11,12,13,14,15])
y = np.array([23,20,20,24,23,21,20])
plt.plot(x, y, color = 'black')
plt.title("湿度時間帯別")
plt.xlabel("hour")
plt.ylabel("temperature")
````
